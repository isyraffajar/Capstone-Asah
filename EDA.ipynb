```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Dataset/Retail_Transactions_Dataset.csv")
df.head()
```

```python
# Visualisasi distribusi
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Distribusi Total_Items
sns.histplot(df['Total_Items'], ax=axes[0,0], bins=50)
axes[0,0].set_title('Distribution of Total Items')

# Distribusi Total_Cost
sns.histplot(df[df['Total_Cost'] < 1000]['Total_Cost'], ax=axes[0,1], bins=50)
axes[0,1].set_title('Distribution of Total Cost (< 1000)')

# Payment Method distribution
df['Payment_Method'].value_counts().plot(kind='bar', ax=axes[0,2])
axes[0,2].set_title('Payment Method Distribution')

# Store Type distribution
df['Store_Type'].value_counts().plot(kind='bar', ax=axes[1,0])
axes[1,0].set_title('Store Type Distribution')

# Customer Category
df['Customer_Category'].value_counts().plot(kind='bar', ax=axes[1,1])
axes[1,1].set_title('Customer Category Distribution')

# Season distribution
df['Season'].value_counts().plot(kind='pie', ax=axes[1,2])
axes[1,2].set_title('Season Distribution')

plt.tight_layout()
plt.show()
```
