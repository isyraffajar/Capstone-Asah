```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Dataset/Retail_Transactions_Dataset.csv")
df.head()
```

```python
# Visualisasi distribusi
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Distribusi Total_Items
sns.histplot(df['Total_Items'], ax=axes[0,0], bins=50)
axes[0,0].set_title('Distribution of Total Items')

# Distribusi Total_Cost
sns.histplot(df[df['Total_Cost'] < 1000]['Total_Cost'], ax=axes[0,1], bins=50)
axes[0,1].set_title('Distribution of Total Cost (< 1000)')

# Payment Method distribution
df['Payment_Method'].value_counts().plot(kind='bar', ax=axes[0,2])
axes[0,2].set_title('Payment Method Distribution')

# Store Type distribution
df['Store_Type'].value_counts().plot(kind='bar', ax=axes[1,0])
axes[1,0].set_title('Store Type Distribution')

# Customer Category
df['Customer_Category'].value_counts().plot(kind='bar', ax=axes[1,1])
axes[1,1].set_title('Customer Category Distribution')

# Season distribution
df['Season'].value_counts().plot(kind='pie', ax=axes[1,2])
axes[1,2].set_title('Season Distribution')

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("../Dataset/kz.csv")

fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# 1. Distribusi Price
sns.histplot(df[df['price'] < 500]['price'], bins=50, ax=axes[0,0])
axes[0,0].set_title('Distribution of Price (< 500)')

# 2. Jumlah transaksi per brand
df['brand'].value_counts().head(20).plot(kind='bar', ax=axes[0,1])
axes[0,1].set_title('Top 20 Brands by Transaction Count')

# 3. Jumlah transaksi per category_code
df['category_code'].value_counts().head(20).plot(kind='bar', ax=axes[0,2])
axes[0,2].set_title('Top 20 Categories')

# 4. Jumlah transaksi per category_id
df['category_id'].value_counts().head(20).plot(kind='bar', ax=axes[1,0])
axes[1,0].set_title('Top 20 Category IDs')

# 5. Jumlah transaksi per user
df['user_id'].value_counts().head(20).plot(kind='bar', ax=axes[1,1])
axes[1,1].set_title('Top 20 Users by Number of Transactions')

# 6. Distribusi waktu transaksi (jam)
df['event_time'] = pd.to_datetime(df['event_time'])
df['hour'] = df['event_time'].dt.hour
sns.countplot(x=df['hour'], ax=axes[1,2])
axes[1,2].set_title('Transaction Count by Hour')

plt.tight_layout()
plt.show()
```
